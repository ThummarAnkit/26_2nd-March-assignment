# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

__Ans. :__<br>

__Matplotlib__ is a Python data visualization library used to create static, interactive, and animated visualizations in Python. It is a powerful and flexible tool that allows users to create a wide range of charts, graphs, and plots for data analysis and presentation.<br>

__Matplotlib is used for__ a variety of tasks, including data exploration, presentation, and publication. It is widely used in scientific computing, engineering, finance, and data science, among other fields.

The Pyplot module of Matplotlib is a collection of functions that provide a convenient interface for creating a variety of plots. Here are five plots that can be plotted using the Pyplot module of Matplotlib:

__1. Line plot:__ A line plot is a basic plot that shows the relationship between two variables using lines.

__2. Scatter plot:__ A scatter plot is a plot that uses dots to represent the relationship between two variables. Each dot represents an observation in the data.

__3. Histogram:__ A histogram is a plot that shows the distribution of a variable by dividing it into a set of bins and counting the number of observations that fall into each bin.

__4. Bar plot:__ A bar plot is a plot that shows the relationship between a categorical variable and a numerical variable using bars.

__5. Pie chart:__ A pie chart is a plot that shows the proportion of each category in a categorical variable as a fraction of a circle.





# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

__Note:__ Also add title, xlabel, and ylabel to the plot.

__Ans. :__<br>

A scatter plot is a type of plot that displays the relationship between two numerical variables. In a scatter plot, each point on the plot represents the values of the two variables, with one variable plotted along the x-axis and the other plotted along the y-axis.

```python
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("this is a X-axis")
plt.ylabel("this is a y-axis")
plt.title("this is a X and y")
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

__Ans. :__<br>

The __subplot()__ function in Matplotlib is used to create multiple plots on the same figure. It takes three arguments: the number of rows, the number of columns, and the plot number. By specifying the plot number, we can create multiple plots in a grid-like layout.

```python
import matplotlib.pyplot as plt
import numpy as np

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of plots
plt.figure(figsize=(8,14))
plt.subplot(4, 1, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(4, 1, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(4, 1, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(4, 1, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Display the plots
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

__Ans. :__<br>

A __bar plot__ is a type of plot that represents categorical data with rectangular bars, where the height or length of the bars corresponds to the magnitude of the data.

__Bar plots__ are commonly used to display and compare the counts, frequencies, or proportions of different categories or groups. They are particularly useful when the data can be naturally separated into categories or when the data is discrete rather than continuous.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

__Ans. :__<br>

A __box plot__, also known as a box-and-whisker plot, is a type of plot that displays the distribution of a dataset. The plot consists of a box that represents the middle 50% of the data (the interquartile range or IQR), with a line inside the box that represents the median value. Whiskers extend from the box to the minimum and maximum values of the data, excluding outliers.

__Box plots__ are useful for identifying the range, spread, skewness, and symmetry of the data. They are commonly used to compare the distributions of different groups or to identify outliers or unusual observations.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data =[box1,box2]
plt.boxplot(data, patch_artist = True, notch ='True', vert = 0)
plt.show()
```
